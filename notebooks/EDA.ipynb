# Load libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
# Load the cleaned dataset
df = pd.read_csv('../data/clean/cleaned_data2.csv')

df.head()
```

# EDA

```python
# Import the required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Define the colors for better visuals
colors = {
    'Anxiety': '#FFAA00', 
    'Burnout': '#0077BB', 
    'Depression': '#EE7733', 
    'None': '#33AA55',
    'Hybrid': '#FFAA00', 
    'Remote': '#0077BB', 
    'Onsite': '#EE7733'
}

# Plot Work_Location vs. Mental_Health_Condition
plt.figure(figsize=(6, 4))
# Define a palette with 3 colors (assuming 3 unique mental health conditions)
sns.countplot(x='Work_Location', hue='Mental_Health_Condition', data=df, palette=sns.color_palette('muted', n_colors=3))
plt.title('Mental Health Conditions across Work Locations')
plt.xlabel('Work Location')
plt.ylabel('Count')
plt.legend(title='Mental Health Condition')
plt.show()

# Plot Stress_Level vs. Job_Role
plt.figure(figsize=(8, 4))
# Adjust the palette to match the number of unique stress levels
sns.countplot(x='Job_Role', hue='Stress_Level', data=df, palette=sns.color_palette('muted', n_colors=3))
plt.title('Stress Levels across Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plot Work_Location vs. Productivity_Change
plt.figure(figsize=(6, 4))
# Adjust the palette for productivity change categories (replace 3 with the number of unique values)
sns.countplot(x='Work_Location', hue='Productivity_Change', data=df, palette=sns.color_palette('muted', n_colors=3))
plt.title('Productivity Change across Work Locations')
plt.xlabel('Work Location')
plt.ylabel('Count')
plt.legend(title='Productivity Change')
plt.show()

```

```python
# Violin plot for Physical Activity across Work Locations with hue assigned
plt.figure(figsize=(8, 4))
sns.violinplot(data=df, x='Work_Location', y='Physical_Activity', hue='Work_Location', palette='Set2', legend=False)
plt.title('Physical Activity across Work Locations')
plt.xlabel('Work Location')
plt.ylabel('Physical Activity Frequency')
plt.show()



from scipy.stats import chi2_contingency

# Create a contingency table for Work Location and Physical Activity
contingency_table_pa = pd.crosstab(df['Work_Location'], df['Physical_Activity'])

# Perform the chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table_pa)

print(f"Chi-Square Test p-value (Work_Location vs. Physical_Activity): {p_val}")
```

```python
# Violin plot for Sleep Quality across Work Locations with palette adjustment
plt.figure(figsize=(8, 4))
sns.violinplot(data=df, x='Work_Location', y='Sleep_Quality', hue='Work_Location', palette='Set2', legend=False)
plt.title('Sleep Quality across Work Locations')
plt.xlabel('Work Location')
plt.ylabel('Sleep Quality')
plt.show()


# Create a contingency table for Work Location and Sleep Quality
contingency_table_sq = pd.crosstab(df['Work_Location'], df['Sleep_Quality'])

# Perform the chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table_sq)

print(f"Chi-Square Test p-value (Work_Location vs. Sleep_Quality): {p_val}")
```

```python
# Violin plot for Social Isolation across Work Locations with palette adjustment
plt.figure(figsize=(8, 4))
sns.violinplot(data=df, x='Work_Location', y='Social_Isolation_Rating', hue='Work_Location', palette='Set2', legend=False)
plt.title('Social Isolation across Work Locations')
plt.xlabel('Work Location')
plt.ylabel('Social Isolation Rating (Higher means more isolated)')
plt.show()


# Create a contingency table for Work Location and Social Isolation Rating
contingency_table_si = pd.crosstab(df['Work_Location'], df['Social_Isolation_Rating'])

# Perform the chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table_si)

print(f"Chi-Square Test p-value (Work_Location vs. Social_Isolation_Rating): {p_val}")
```

```python
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Mental Health Conditions by Work Location
mental_health_counts = df.groupby('Work_Location')['Mental_Health_Condition'].count()
axes[0, 0].bar(mental_health_counts.index, mental_health_counts.values)
axes[0, 0].set_title('Mental Health Conditions by Work Location')

# Plot 2: Stress Levels by Job Role (keeping job roles as categorical)
stress_levels = df.groupby('Job_Role')['Stress_Level'].mean()
axes[0, 1].bar(stress_levels.index, stress_levels.values)
axes[0, 1].set_title('Stress Levels by Job Role')

# Plot 3: Productivity Change by Work Location
productivity_change_counts = df.groupby('Work_Location')['Productivity_Change'].value_counts().unstack()
productivity_change_counts.plot(kind='bar', stacked=True, ax=axes[1, 0])
axes[1, 0].set_title('Productivity Change by Work Location')

# Plot 4: Work-Life Balance by Job Role (keeping job roles categorical)
work_life_balance = df.groupby('Job_Role')['Work_Life_Balance_Rating'].mean()
axes[1, 1].bar(work_life_balance.index, work_life_balance.values)
axes[1, 1].set_title('Work-Life Balance by Job Role')

plt.tight_layout()
plt.show()
```
